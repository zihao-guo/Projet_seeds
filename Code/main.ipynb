```python
# Initialize data
init_alpha0 = 0
init_alpha1 = 0
init_alpha2 = 0
init_alpha12 = 0
init_sigma = 1/10
init_b = [0] * 21

alpha0, alpha1, alpha2, alpha12, b, sigma = init_alpha0, init_alpha1, init_alpha2, init_alpha12, init_b, init_sigma

x1 = data['x1'].values
x2 = data['x2'].values
r = data['r'].values
n = data['n'].values
```

## 2. Modeling

```python
def sigmoid(x):
    """定义sigmoid函数"""
    return 1 / (1 + np.exp(-x))

def lh(alpha0, alpha1, alpha2, alpha12, b, r, x1, x2, n, K):
    a = 0  # 初始化似然值
    for i in range(K):
        a += r[i] * np.log(p(alpha0, alpha1, alpha2, alpha12, x1[i], x2[i], b[i])) + (n[i] - r[i]) * np.log(1 - p(alpha0, alpha1, alpha2, alpha12, x1[i], x2[i], b[i]))
    return a

def p(alpha0, alpha1, alpha2, alpha12, x1, x2, b):
    """定义概率密度函数p"""
    return sigmoid(alpha0 + alpha1 * x1 + alpha2 * x2 + alpha12 * x1 * x2 + b)
```

```python
nchain = 10000
N = 21
prop_sd = [0.5,0.5,0.5,0.5,0.01]
```

```python
chain = np.zeros((nchain+1,5)) # alpha0,alpha1,alpha2,alpha12,sigma
chain_b = np.zeros((nchain+1, N))
#Initialisation
chain[0] = [alpha0, alpha1, alpha2, alpha12, sigma]
chain_b[0] = init_b
```

```python
for i in range(nchain):
    chain[i+1] = chain[i]
    
    #Mise à jour alpha0
    prop = chain[i+1][0] + prop_sd[0]*np.random.normal() #proposition alpha0
    
    top = -prop**2/(2*10**6) + lh(prop, chain[i+1][1], chain[i+1][2], chain[i+1][3], chain_b[i+1], r, x1, x2, n, N)
    bottom = -alpha0**2/(2*10**6) + lh(chain[i+1][0], chain[i+1][1], chain[i+1][2], chain[i+1][3], chain_b[i+1], r, x1, x2, n, N) #vraisemblance
    
    prob = np.exp(top - bottom)
    if np.random.uniform() < prob:
        chain[i+1][0] = prop
        
        
    #Mise à jour alpha1
    prop = chain[i+1][1] + prop_sd[1]*np.random.normal() #proposition alpha0
    
    top = -prop**2/(2*10**6) + lh(chain[i+1][0], prop, chain[i+1][2], chain[i+1][3], chain_b[i+1], r, x1, x2, n, N)
    bottom = -alpha0**2/(2*10**6) + lh(chain[i+1][0], chain[i+1][1], chain[i+1][2], chain[i+1][3], chain_b[i+1], r, x1, x2, n, N) #vraisemblance
    
    prob = np.exp(top - bottom)
    if np.random.uniform() < prob:
        chain[i+1][1] = prop
     
    
    #Mise à jour alpha2
    prop = chain[i+1][2] + prop_sd[2]*np.random.normal() #proposition alpha0
    
    top = -prop**2/(2*10**6) + lh(chain[i+1][0], chain[i+1][1], prop, chain[i+1][3], chain_b[i+1], r, x1, x2, n, N)
    bottom = -alpha0**2/(2*10**6) + lh(chain[i+1][0], chain[i+1][1], chain[i+1][2], chain[i+1][3], chain_b[i+1], r, x1, x2, n, N) #vraisemblance
    
    prob = np.exp(top - bottom)
    if np.random.uniform() < prob:
        chain[i+1][2] = prop
        
        
    #Mise à jour alpha12
    prop = chain[i+1][3] + prop_sd[3]*np.random.normal() #proposition alpha0
    
    top = -prop**2/(2*10**6) + lh(chain[i+1][0], chain[i+1][1], chain[i+1][2], prop, chain_b[i+1], r, x1, x2, n, N)
    bottom = -alpha0**2/(2*10**6) + lh(chain[i+1][0], chain[i+1][1], chain[i+1][2], chain[i+1][3], chain_b[i+1], r, x1, x2, n, N) #vraisemblance
    
    prob = np.exp(top - bottom)
    if np.random.uniform() < prob:
        chain[i+1][3] = prop
        
    #Mise à jour sigma
    shape = 0.01 + N/2,
    scale = 0.001 + 0.5 * sum(chain_b[i]**2) #？？？
    tau = gamma.rvs(shape, scale)
    chain[i+1][4] = 1/np.sqrt(tau)
    
    #Mise à jour b
    chain_b[i+1]=chain_b[i]
    for j in range(N):
        Xp = chain_b[i+1][j] + prop_sd[4]*np.random.normal() #proposition b
        top0 = -Xp**2/(2*chain[i+1][4]**2) #loi a priori b
        top1 = r[j]*np.log(p(chain[i+1][0], chain[i+1][1], x1[j], chain[i+1][2], x2[j], chain[i+1][3], Xp)) + (n[j]-r[j])*np.log(1-p(chain[i+1][0], chain[i+1][1], x1[j], chain[i+1][2], x2[j], chain[i+1][3], Xp))
        top = top0 + top1
        bottom0 = -chain_b[i+1][j]**2/(2*chain[i+1][4]**2) #loi a priori b
        bottom1 = r[j]*np.log(p(chain[i+1][0], chain[i+1][1], x1[j], chain[i+1][2], x2[j], chain[i+1][3], chain_b[i+1][j])) + (n[j]-r[j])*np.log(1-p(chain[i+1][0], chain[i+1][1], x1[j], chain[i+1][2], x2[j], chain[i+1][3], chain_b[i+1][j]))
        bottom = bottom0 + bottom1
        prob = np.exp(top - bottom)
        if np.random.uniform() < prob:
            chain_b[i+1][j] = Xp
```

```python
variables = ["alpha0", "alpha1", "alpha2", "alpha12", "sigma"]
values = [np.round(np.mean(chain[:, i]), 3) for i in range(len(variables))]
for variable, value in zip(variables, values):
    print(f"{variable} a posteriori: {value}")
```

```python
variables = ["alpha0", "alpha1", "alpha2", "alpha12", "sigma"]
values = [np.round(np.std(chain[:, i]), 3) for i in range(len(variables))]
for variable, value in zip(variables, values):
    print(f"{variable} a posteriori: {value}")
```
